# Create Spark Session


```python

import pyspark

from pyspark.sql import SparkSession
s=SparkSession.builder.appName('ecommerce').getOrCreate()
```

```python
s
```

## Read csv file

```python

# header=True → first row contains column names
# inferSchema=True → Spark automatically detects data types

df = s.read.csv(
    "ecommerce_dirty_dataset.csv",  # file name
    header=True,        # Use first row as column names
    inferSchema=True    # Automatically detect column data types
)
```

## Add column

## Rename the column




```python
df=df.withColumnRenamed('price with GST','Total price')
```

## Drop column



```python
## droping the "price" column

df.drop('price').show()
```

```python
## shape
from pyspark.sql.functions import col

rows = df.count()
cols = len(df.columns)

print(rows, cols)
```

```python
df.filter(col("city").isNull()).count()
```

## Null value handling

```python
# drop all null values  if any null value contain in  row


df.na.drop(how='any').show()
```

```python
# drop the row if only the all colums are null in a row

df.na.drop(how='all').show()
```

```python
# drop the null value in specific column like 'name'

df.na.drop(subset='customer_name',how='all').show()
```

## Filling the missing values

```python
# relace with a specic value in entaire dataframe . eg- 'not found' - only fill the categorical column

df.na.fill('not found').show()
```

```python
# fill the null values of specific columns . eq - fill with 25 in the colums "price" and "quantity"

df.na.fill(25,['quantity','price']).show()

```

## Filling with mean 

```python
numeric_cols=[col for col,dtype in df.dtypes if dtype in ('int','double','float')]

df.select(numeric_cols).show()
```

```python
numeric_cols
```

```python
# filling with mean 

from pyspark.ml.feature import Imputer

imputer=Imputer(
    inputCols=['order_id', 'quantity', 'price', 'Total price'],
    outputCols=['{}_imputed'.format(c)for c in ['order_id', 'quantity', 'price', 'Total price']]

).setStrategy('mode')
```

```python
imputer.fit(df).transform(df).show()
```

## Filter options


```python
df.filter("'Total price'>100")
```

```python
df.filter("`Total price` > 1000").select('customer_name','price').show()
```

```python
df.filter("`Total price` > 1000").select('Total price','Product').show()
```

```python
df.filter(df['Total price']>=1700).show()
```

## Group by

```python
df.groupBy('customer_name').sum()
```

```python
df.groupBy('customer_name').sum().show()
```

```python
df.groupBy('customer_name').sum('price').show()
```

```python
df.groupBy('customer_name').max('price').show()
```

```python
df.groupBy('product').sum('price').show()
```

```python
df.groupBy('product').mean('price').show()
```

```python
df.groupBy('product').count().show()
```

```python
df.agg({'price':'sum'}).show()
```

```python
df.groupBy('product').agg({'price':'sum','price':'max'}).show()
```

```python
from pyspark.sql.functions import sum, max
df.groupBy('product').agg(sum(
    'price')
    ,max('price')
).show()
```

```python


df.groupBy('product').agg(
    sum('price').alias('total_price'),
    max('price')
).show()
```
